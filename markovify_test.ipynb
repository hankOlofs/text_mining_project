```python
amazon_raw = pd.read_csv('train.csv', header=None)
```

```python
amazon = amazon_raw.iloc[0:100000,:]
amazon.columns = ['polarity', 'title', 'text']
```

```python
amazon.polarity.replace({1:0, 2:1}, inplace=True)
amazon.head(20)
```

```python
amazon.polarity.value_counts()
```

```python
amazon_text = '\n'.join([i for i in amazon['text'].apply(str)])
```

```python
amazon_text_model = markovify.NewlineText(amazon_text)
```

```python
for i in range(5):
    print(amazon_text_model.make_sentence(), '\n')
```

```python
import pickle
pickle.dump(amazon_text_model, open('amazon_model.pkl', 'wb'))
```

```python
# load pickled model
amazon_text_model = pickle.load(open('amazon_model.pkl', 'rb'))
```
